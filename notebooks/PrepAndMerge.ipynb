# Data Visualization 

```python
import collections
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from matplotlib import rcParams
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
```

```python
dataset = pd.read_csv('data/IBW.csv', encoding='latin-1')
all_headlines = ' '.join(dataset['Phrase'].str.lower())
```

```python
stopwords = STOPWORDS
stopwords.add('teri')
stopwords.add('ko')
stopwords.add('ki')
stopwords.add('maa')
stopwords.add('ke')
stopwords.add('me')
stopwords.add('ma')
stopwords.add('ka')
stopwords.add('mein')
stopwords.add('baal')
stopwords.add('hai')
stopwords.add('mera')
stopwords.add('bhen')
stopwords.add('na')
stopwords.add('le')
stopwords.add('lo')
stopwords.add('tor')
stopwords.add('da')
stopwords.add('bal')
stopwords.add('tere')
stopwords.add('ni')
stopwords.add('bahen')
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white", max_words=1000).generate(all_headlines)
rcParams['figure.figsize'] = 10, 20
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
filtered_words = [word for word in all_headlines.split() if word not in stopwords]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top words in the headlines vs their count')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)
```

```python
import numpy as np

REPLACE = {'a': '@', 'i': '*', 'o': '*', 'u': '*', 'v': '*', 
 'l': '1', 'e': '*', 's': '$', 't': '7'}

from itertools import product

def Leet2Combos(word):
    possibles = []
    for l in word.lower():
        ll = REPLACE.get(l, l)
        possibles.append( (l,) if ll == l else (l, ll) )
    return [ ''.join(t) for t in product(*possibles) ]
s="""Chodon Boiraagi
Chodon Idli
chood
chudir bhai
Chudir Bhai"""

words = s.split('\n')
print(words)
lst=[]
# ['Chodon Boiraagi', 'Chodon Idli', 'chood']
for word in words:
    lst.append(Leet2Combos(word))
    k = pd.DataFrame(lst)
```
